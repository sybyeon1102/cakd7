# 범주화

````python
def get_category(x):
    cat = ''
    if x <11:
        cat = 'child'
    elif x < 30:
        cat = 'young adult'
    elif x < 60:
        cat = 'adult'
    else:
        cat = 'elderly'
    return cat

df['age_class'] = df.age.apply(lambda x: get_category(x))
df[['age', 'age_class']].head()
    
````

````python
df.head()
````

````python
df['sex'] = df.sex.apply(lambda x: x.upper())

df['sex'].head()
````
